```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('base_prob.csv')
df
```

```python
plt.figure(figsize=(15,5))

plt.hist(df.age, bins=np.arange(0,101,1),rwidth=.8, density=True);
plt.axvline(29,color='k')
plt.axhline(0.0125,color='k')

plt.xlabel('Probabilty/Age',fontsize=16)
plt.ylabel('Count of people',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks([0, 0.005, 0.01, 0.0125, 0.015, 0.02 , 0.025],fontsize=16)
plt.show()
```

## Joint probability distribution

```python
plt.figure(figsize=(15,10))

plt.subplot(2,2,1)
plt.hist(df.MonthlyIncome,bins=30, density=True, orientation='horizontal',rwidth=.8)
plt.xticks(fontsize=16,rotation=45)
plt.yticks(fontsize=16)

plt.subplot(2,2,2)
plt.hist2d(df.age,df.MonthlyIncome,bins=30, density=True)
plt.colorbar()

plt.xlabel('Probabilty/Age',fontsize=16)
plt.ylabel('Count of people',fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.subplot(2,2,4)
plt.hist(df.age,bins=30, density=True, orientation='vertical',rwidth=.8)
plt.xticks(fontsize=16,rotation=45)
plt.yticks(fontsize=16)

plt.show()
```
